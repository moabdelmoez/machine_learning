```python
from sklearn.datasets import load_boston
boston = load_boston()
```

```python
# The boston variable itself is a dictionary, so you can check for its keys using the .keys() method
print(boston.keys())
```

```python
# You can easily check for its shape by using the boston.data.shape attribute,
# which will return the size of the dataset
print(boston.data.shape)

# As you can see it returned (506, 13), that means there are 506 rows of data with 13 columns
```

```python
# if you want to know what the 13 columns are, you can simply use the .feature_names attribute and
# it will return the feature names
print(boston.feature_names)
```

```python
# The description of the dataset is available in the dataset itself
print(boston.DESCR)
```

```python
import pandas as pd

# letâ€™s convert it into a pandas DataFrame
data = pd.DataFrame(boston.data)
data.columns = boston.feature_names
```

```python
data.head()
```

```python
data['PRICE'] = boston.target
```

```python
data.info()
```

```python
data.describe()
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
# Separate the target variable and rest of the variables using .iloc to subset the data
X, y = data.iloc[:, :-1], data.iloc[:, -1]
```

```python
# Now you will convert the dataset into an optimized data structure called Dmatrix that XGBoost supports and 
# gives it acclaimed performance and efficiency gains
data_dmatrix = xgb.DMatrix(data=X, label=y)
```

```python
data_dmatrix
```

```python
"""
Now, you will create the train and test set for cross-validation of the results using the train_test_split function
from sklearn's model_selection module with test_size size equal to 20% of the data. 
Also, to maintain reproducibility of the results, a random_state is also assigned
"""

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)
```

```python
"""
The next step is to instantiate an XGBoost regressor object by calling the XGBRegressor() class from
the XGBoost library with the hyper-parameters passed as arguments. 
For classification problems, you would have used the XGBClassifier() class.
"""
xg_reg = xgb.XGBRegressor(objective='reg:linear', colsample_bytree=0.3, learning_rate=0.1,
                         max_depth=5, alpha=10, n_estimators=10)
```

```python
# Fit the regressor to the training set and make predictions on the test set
xg_reg.fit(X_train, y_train)
preds = xg_reg.predict(X_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, preds))
print("RMSE: %f" % (rmse))
```

```python
params = {"objective": "reg:linear", "colsample_bytree": 0.3, "learning_rate": 0.1,
         "max_depth": 5, "alpha": 10}

cv_results = xgb.cv(dtrain=data_dmatrix, params=params, nfold=3, num_boost_round=50, early_stopping_rounds=10,
                   metrics="rmse", as_pandas=True, seed=123)
```
